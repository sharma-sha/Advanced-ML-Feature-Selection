```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv("prep.csv")
```

```python
df=pd.get_dummies(df,dtype=int,drop_first=True)
```

```python
df
```

```python
print(df.dtypes)
```

```python
# Step 2: Select the first 14 columns as independent variables (X)
X = df.iloc[:, :-1]  # Select all columns except the last one

# Step 3: Select the last column as the dependent variable (y)
y = df.iloc[:, -1]  # Select the last column only
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X.head()
```

```python
y
```

```python
X_train.shape
```

```python
forward_feature_selection = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                     k_features= 6,
                                                     forward=True,
                                                     floating=False,
                                                     verbose=2,
                                                     scoring= "accuracy",
                                                     cv= 5).fit(X_train, y_train)
```
